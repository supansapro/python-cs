```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def average_score(self):
        if len(self.scores) == 0:
            return 0
        return sum(self.scores) / len(self.scores)

    def get_info(self):
        return f"{self.name} - {self.scores}"

student1 = Student("Alice")
student2 = Student("Tong")
student3 = Student("Fa")


student1.add_score(90)
student1.add_score(85)
student1.add_score(88)

student2.add_score(75)
student2.add_score(80)
student2.add_score(70)

student3.add_score(60)
student3.add_score(65)
student3.add_score(70)


students = [student1, student2, student3]

for student in students:
    print(student.get_info())
    print(f"Average score: {student.average_score():.2f}")
    print("-" * 30)

```

```python
class Student:
    def __init__(self, name):
        self.name = name
        self.scores = []

    def add_score(self, score):
        self.scores.append(score)

    def average_score(self):
        return sum(self.scores) / len(self.scores) if self.scores else 0

    def __str__(self):
        return f"{self.name} - {self.scores}"

class Classroom:
    def __init__(self):
        self.students = []

    def add_student(self, student):
        self.students.append(student)

    def insert_student(self, index, student):
        self.students.insert(index, student)

    def find_student(self, name):
        for i, student in enumerate(self.students):
            if student.name == name:
                return i
        return -1

    def remove_student(self, name):
        index = self.find_student(name)
        if index != -1:
            self.students.pop(index)

    def sort_students(self):
        self.students.sort(key=lambda s: s.name)

    def reverse_students(self):
        self.students.reverse()

    def clear_students(self):
        self.students.clear()

    def get_all_students_info(self):
        return [str(student) for student in self.students]


room = Classroom()

def add_students():
    global s1, s2, s3, s4, s5
    s1 = Student("Alice")
    s1.add_score(90)
    s1.add_score(85)
    s1.add_score(70)
    print(f"Alice - {s1.scores} Average: {s1.average_score()}")

    s2 = Student("Bob")
    s2.add_score(60)
    s2.add_score(75)
    s2.add_score(80)
    print(f"Bob - {s2.scores} Average: {s2.average_score()}")

    s3 = Student("Charlie")
    s3.add_score(100)
    s3.add_score(95)
    s3.add_score(90)
    print(f"Charlie - {s3.scores} Average: {s3.average_score()}")

    room.add_student(s1)
    room.add_student(s2)
    room.add_student(s3)

    s4 = Student("David")
    s5 = Student("Eva")
    room.add_student(s4)
    room.add_student(s5)

    print("\nAll students:", room.get_all_students_info())

def insert_student():
    s6 = Student("Frank")  
    room.insert_student(2, s6)
    print("\nAfter Insert:", room.get_all_students_info())

def sort_students():
    room.sort_students()
    print("\nSorted:", room.get_all_students_info())

def reverse_students():
    room.reverse_students()
    print("\nReversed:", room.get_all_students_info())

def find_student(name):
    index = room.find_student(name)
    print(f"\nIndex of {name}:", index)

def remove_student(name):
    room.remove_student(name)
    print("\nAfter remove:", room.get_all_students_info())

def clear_students():
    room.clear_students()
    print("\nAfter clear:", room.get_all_students_info())



add_students()
insert_student()
sort_students()
reverse_students()
find_student("Bob")
remove_student("Bob")
clear_students()
```
